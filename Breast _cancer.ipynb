```python
# Data visualization
data=pd.read_csv("data.csv")
df=pd.DataFrame(data)
df.head()
```

```python
df=df.drop(['id','Unnamed: 32'],axis=1)
df.head()
```

```python
# Finding Correlation between featurs
df1=df.drop(['diagnosis'],axis =1)
corr_matrix=df1.corr().abs()
corr_matrix
```

```python
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(bool))
upper_tri.head()
```

```python
# List of  corelated features 
To_drop=[column for column in upper_tri.columns if any(upper_tri[column]>0.95)]
print(To_drop)
```

```python
# Deleting the correlated features
for i in To_drop:
    df=df.drop([i],axis=1)
df.head()
```

```python
# replacing the categorical variables with numbers
from sklearn.preprocessing import LabelEncoder
num=LabelEncoder()
df['diagnosis']=num.fit_transform(df['diagnosis'].astype('str'))
```

```python
# spliting data set into features and target 
x=df.drop(columns=['diagnosis'])
y=df.diagnosis
```

```python
# Importing  Machine Learning Libaries
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
```

```python
X_,x_,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
# creating the list of machine learning model
models={}
models["LogisticRegression"]=LogisticRegression(random_state=0)
models["RandomForestClassifier"]=RandomForestClassifier(n_estimators=10, random_state=0)
models["KNeighborsClassifier"]=KNeighborsClassifier()
models["SupportVectorMachine"]=SVC(C=0.1,kernel='linear',gamma=1)
```

```python
# Function for classification,accuracy,confusion matrix ,precision,recall, F1score etc
def Model_Building(name,model,X_train,X_test,Y_train,Y_test):
    model.fit(X_train,Y_train)
    pred=model.predict(X_test)
    accuracy=accuracy_score(pred,Y_test)
    print("Predictions for {} ".format(name))
    print("Confusion Marix:")
    con_mat=confusion_matrix(pred,Y_test)
    plot=sns.heatmap(con_mat,annot=True,cmap='Blues')
    plot.set_xlabel('Predicted')
    plot.set_ylabel('Actual')
    plt.show()
    print("Accuracy:   {}".format(accuracy))
    precision=precision_score(pred,Y_test)
    recall =recall_score(pred,Y_test)
    F1_score=2*precision*recall/(precision+recall)
    print("precision: {}".format(precision))
    print("Recall: {}".format(recall))
    print("F1_score: {}".format(F1_score))
    print('\n')
    
    
for name,model in models.items():
    Model_Building(name,model,X_train,X_test,y_train,y_test)
    
    
```
